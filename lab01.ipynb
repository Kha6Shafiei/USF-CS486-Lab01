```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load the dataset



```python
df = pd.read_csv('fifa_2019.csv', index_col=0)
df

```

# Highest Salary

```python
salary = df.Wage.str.strip("€K").astype(int).max()
salary 
```

```python
df.Wage
```

# Average age of player

```python
ave_age = df['Age'].mean()
ave_age
```

# name of oldest player

```python
name_oldest_player = df[df.Age == df.Age.max()]
name_oldest_player
```

# Histogram of Agility

```python
histo = df.Reactions.hist(bins=10)

histo.set_title('Players Reaction in FIFA 2019')
histo.set_xlabel('Reactions')
histo.set_ylabel('Number of players')

plt.show()
```

# Correlation

```python
correlation = df.corr()
correlation
```

```python
import seaborn as sns
sns.heatmap(correlation)
```

```python


plt.scatter(x=df.BallControl, y=df.Finishing)
plt.xlabel('Ball Control')
plt.ylabel('Finishing')
plt.title('Players with better Ball contorl have better finishing')
plt.show()
```

```python
plt.figure().add_axes([0, 0, 1, 1])
plt.bar(df.Age, df.Reactions)
plt.xlabel('Age')
plt.ylabel('Reactions')
plt.title('Reaction By Age')
plt.show()
```
